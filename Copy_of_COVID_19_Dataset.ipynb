```python
# Importing necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Main dataset
df = pd.read_csv('/content/covid_19_data.csv')

# Load datasets
confirmed = pd.read_csv('/content/time_series_covid_19_confirmed.csv')
deaths = pd.read_csv('/content/time_series_covid_19_deaths.csv')
recovered = pd.read_csv('/content/time_series_covid_19_recovered.csv')

# Drop unnecessary columns
confirmed_time_series = confirmed.drop(columns=['Province/State', 'Country/Region', 'Lat', 'Long']).sum(axis=0)
deaths_time_series = deaths.drop(columns=['Province/State', 'Country/Region', 'Lat', 'Long']).sum(axis=0)
recovered_time_series = recovered.drop(columns=['Province/State', 'Country/Region', 'Lat', 'Long']).sum(axis=0)

# Extract dates for plotting
dates = confirmed.columns[4:]

# Ensure all time series data have the same length as the dates
confirmed_time_series = confirmed_time_series[:len(dates)]
deaths_time_series = deaths_time_series[:len(dates)]
recovered_time_series = recovered_time_series[:len(dates)]

# Plotting the data with matching lengths
plt.figure(figsize=(12,6))
plt.plot(dates, confirmed_time_series, label='Confirmed', color='blue')
plt.plot(dates, deaths_time_series, label='Deaths', color='red')
plt.plot(dates, recovered_time_series, label='Recovered', color='green')
plt.title('Worldwide Covid-19 Trend')
plt.xlabel('Date')
plt.ylabel('Cases')
plt.legend()
plt.xticks(rotation=90)
plt.show()

# Worldwide total confirmed cases, deaths, and recovered
total_confirmed = confirmed_time_series.iloc[-1]
total_deaths = deaths_time_series.iloc[-1]
total_recovered = recovered_time_series.iloc[-1]

print(f"Total Confirmed: {total_confirmed}")
print(f"Total Deaths: {total_deaths}")
print(f"Total Recovered: {total_recovered}")

# Get the latest date for each dataset
latest_date = dates[-1]
print(f"Latest date: {latest_date}")

# Confirmed cases, deaths, and recovered cases for the latest date
latest_confirmed = confirmed_time_series.iloc[-1]
latest_deaths = deaths_time_series.iloc[-1]
latest_recovered = recovered_time_series.iloc[-1]

print(f"Confirmed cases on {latest_date}: {latest_confirmed}")
print(f"Deaths on {latest_date}: {latest_deaths}")
print(f"Recovered on {latest_date}: {latest_recovered}")

# Plot daily new cases (Confirmed, Deaths, Recovered)
new_confirmed = confirmed_time_series.diff().fillna(0)
new_deaths = deaths_time_series.diff().fillna(0)
new_recovered = recovered_time_series.diff().fillna(0)
plt.figure(figsize=(12,6))
plt.plot(dates, new_confirmed, label='New Confirmed', color='blue')
plt.plot(dates, new_deaths, label='New Deaths', color='red')
plt.plot(dates, new_recovered, label='New Recovered', color='green')
plt.title('Daily Covid-19 Cases (Confirmed, Deaths, Recovered)')
plt.xlabel('Date')
plt.ylabel('New Cases')
plt.legend()
plt.xticks(rotation=90)
plt.show()
```
